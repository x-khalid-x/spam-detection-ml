```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import re
```

```python
df = pd.read_csv("../data/spam.csv")
df.head()
```

```python
stop_words = set(stopwords.words("english"))
stemmer = PorterStemmer()
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r"\W", " ", text)
    text = re.sub(r"\d", "", text)
    words = [
        stemmer.stem(word)
        for word in text.split()
        if word not in stop_words
    ]
    return " ".join(words)
```

```python
df["Message_clean"] = df["Message"].apply(clean_text)
df.head()
```

```python
df[["Message", "Message_clean"]].head(10)
```

```python
df_clean = df[["Category", "Message_clean"]]
df_clean = df_clean.rename(columns={"Message_clean": "Message"})
df_clean.head()
```

```python
df_clean.to_csv("../data/spam_clean.csv", index=False)
print("Dataset nettoyé sauvegardé")
```
