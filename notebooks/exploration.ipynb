```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("default")
```

```python
df = pd.read_csv("../data/spam.csv")
df.head()
```

```python
df.info()
```

```python
print("Nombre de messages :", df.shape[0])
print("Nombre de colonnes :", df.shape[1])
```

```python
df["Category"].value_counts()
```

```python
sns.countplot(x="Category", data=df)
plt.title("RÃ©partition des messages Spam / Non-Spam")
plt.show()
```

```python
df["message_length"] = df["Message"].apply(len)
df[["Category", "message_length"]].head()
```

```python
df.groupby("Category")["message_length"].describe()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(data=df, x="message_length", hue="Category", bins=50, kde=True)
plt.title("Distribution de la longueur des messages")
plt.show()
```
